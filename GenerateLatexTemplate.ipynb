```python
filepath=repofolder+"_data/byword.csv"
df=pd.read_csv(filepath,sep="\t").fillna("")
df["extendedStrongNumber"]=df["extendedStrongNumber"].apply(lambda x: x.replace("_","-"))
for i in df.loc[df["separ"]=="־"]["separ"]: i=""
for i in df.loc[df["separ"]==" "]["separ"]: i=""
```

```python
dic_trans = dict(zip("0123456789","AZERTYUIOP"))
```

```python
def transform(s):
    return "E"+"".join([dic_trans[letter] for letter in s[1:]])
def transform2(s):
    return "".join([dic_trans[letter] for letter in s])
```

```python
transform("E70003")
```

```python
def separator(k):
    if k=="": return k
    if k==" ": return ""
    return "\\separ{{{}}}".format(k)
```

```python
df["latex1"] = df.apply(lambda x: "$\\frac{{\\txnm{{\\{}}}}}{{\\frac{{\\txttt{{{}}}}}{{\\frac{{\\txn{{{}}}}}{{\\txn{{{}}}}}}}}}$ {}".format(transform(x["lexemeID"]),x["morphology"],x["lexemeID"],x["extendedStrongNumber"],separator(x["separ"])),axis=1)
```

```python
temp=df.groupby(["lexemeID","extendedStrongNumber"])["extendedStrongNumber","WLCverse"].agg({'extendedStrongNumber': 'min' , 'WLCverse':'count'})\
.groupby(["lexemeID"]).agg(lambda x: list(x))
```

```python
def argmax(lst):
    return lst.index(max(lst))
```

```python
temp["argmax"]=temp["WLCverse"].apply(argmax)
temp["mainentry"]=temp.apply(lambda x: x["extendedStrongNumber"][x["argmax"]], axis=1)
temp["percentlost"]=temp.apply(lambda x: 1-x["WLCverse"][x["argmax"]]/sum(x["WLCverse"]), axis=1)
temp["numberlost"]=temp.apply(lambda x: sum(x["WLCverse"])-x["WLCverse"][x["argmax"]], axis=1)
temp["totalocc"]=temp.apply(lambda x: sum(x["WLCverse"]), axis=1)
```

```python
temp[(temp["numberlost"]>0) & (temp["percentlost"]>.01)]
```
